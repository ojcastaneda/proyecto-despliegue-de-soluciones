```python
from pandas import read_csv
from matplotlib.pyplot import subplots, savefig, show, tight_layout


def plot_logs(model_name: str):
    cls_df = read_csv(f"../train_logs/{model_name}_classification.csv")
    det_df = read_csv(f"../train_logs/{model_name}_detection.csv")
    best_cls_step = cls_df.loc[cls_df["macro_f1"].idxmax(), "step"]
    best_det_step = det_df.loc[det_df["macro_f1"].idxmax(), "step"]
    _, axes = subplots(1, 4, figsize=(20, 5), sharex=False)
    axes[0].plot(cls_df["step"], cls_df["train_loss"], label="Loss", color="black")
    axes[0].axvline(x=best_cls_step, color='red', linestyle='--')
    axes[0].set_title("Classification Loss")
    axes[0].set_xlabel("Step")
    axes[0].set_ylabel("Loss")
    axes[1].plot(cls_df["step"], cls_df["macro_f1"], label="Macro F1", color='green')
    axes[1].plot(cls_df["step"], cls_df["accuracy"], label="Accuracy", color='blue')
    axes[1].axvline(x=best_cls_step, color='red', linestyle='--')
    axes[1].set_title("Classification F1 / Accuracy")
    axes[1].set_xlabel("Step")
    axes[1].legend()
    axes[2].plot(det_df["step"], det_df["train_loss"], label="Loss", color="black")
    axes[2].axvline(x=best_det_step, color='red', linestyle='--')
    axes[2].set_title("Detection Loss")
    axes[2].set_xlabel("Step")
    axes[3].plot(det_df["step"], det_df["macro_f1"], label="Macro F1", color='green')
    axes[3].plot(det_df["step"], det_df["accuracy"], label="Accuracy", color='blue')
    axes[3].axvline(x=best_det_step, color='red', linestyle='--')
    axes[3].set_title("Detection F1 / Accuracy")
    axes[3].set_xlabel("Step")
    axes[3].legend()
    tight_layout()
    savefig(f"../train_logs/{model_name}.png", dpi=300)
    show()
```

```python
plot_logs("bloom")
```
